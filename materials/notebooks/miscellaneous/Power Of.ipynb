```
def power_of(n: int, pow: int) -> int:
  """Obviously redundant with Python syntax n**pow, but fun.
  Challenge... can it be done recursively with only two arguments?"""
  val = 1
  for i in range(pow):
    val *= n
  return val

def prompt_int(looking_for: str) -> int:
  """Prompt for and return an integer, asking for the given noun phrase."""

  def _check_int(s: str) -> bool:
    """Return True iff the given input is an integer. Hackish implementation!"""
    try:
      int(s)
      return True
    except:
      return False

  prompt = f'Enter {looking_for}: '
  n = input(prompt)
  while not _check_int(n):
    print('Not a valid integer!\n')
    n = input(prompt)
  
  return int(n)

# Demo
n = prompt_int('a base number')
pow = prompt_int('a power to raise it to')
print(f'{n}^{pow} = {power_of(n, pow)}')
```
