```
class UserQuitException(Exception):
  pass

def find_pair(mult_to: int, add_to: int) -> 'tuple':
  for i in range(abs(mult_to) // 2 + 1):
    low = i + 1
    if mult_to % low:
      continue
    
    high = abs(mult_to) // low

    if mult_to < 0:
      if add_to < 0:
        high = -high
      else:
        low = -low

    elif add_to < 0:
      low = -low
      high = -high
      
    if low * high == mult_to:
      if low + high == add_to:
        return low, high
    
  return tuple()

def has_quit(s: str, can_quit: bool) -> bool:
  return can_quit and not s.strip()

def quit() -> None:
  raise UserQuitException('You decided to quit')

def prompt_int(looking_for: str, can_quit: bool=False) -> int:
  """Prompt for and return an integer, asking for the given noun phrase."""

  def _check_int(s: str) -> bool:
    """Return True iff the given input is an integer. Hackish implementation!"""
    try:
      int(s)
      return True
    except:
      return False

  if can_quit:
    pt = f'Enter {looking_for} (or hit Enter to quit): '
  else:
    pt = f'Enter {looking_for}: '

  n = input(pt)
  while not (_check_int(n) or has_quit(n, can_quit)):
    print('Not a valid integer!\n')
    n = input(pt)

  if has_quit(n, can_quit):
    quit()
  
  return int(n)

while True:
  try:
    mult_to = prompt_int('number to multiply to', True)
    add_to = prompt_int('number to add to', True)
  except UserQuitException:
    print('You decided to quit')
    break

  sol = find_pair(mult_to, add_to)
  if not sol:
    print('Unfactorable!')
  else:
    print(f'{sol[0]} and {sol[1]}')
  print()
```
